```python
import os
os.sys.path
import cv2
import numpy as np

vidcap = cv2.VideoCapture('badapple.mp4')
```

```python
# Criar o vetor com os momentos (em ms) em que são sampleados os frames do vídeo
videoLength_ms = 219*1000
step = 100

samples = np.arange(0, videoLength_ms, step)
```

```python
# Tamanho do "vídeo" mostrado no moodle
dim = (153, 115)

pixel_branco = "&#9633"
pixel_preto = "&#9632"

# Função para gerar uma string referente a um frame
def getFrame(milisec):
    # Colocar o vídeo no tempo dado e obter o frame
    vidcap.set(cv2.CAP_PROP_POS_MSEC, milisec)
    hasFrames,image = vidcap.read()
    
    # Se o frame existe...
    if hasFrames:
        # Início do comando para inserir
        pixel_string = ""
        pixel_string = pixel_string + "setTimeout(function(){ \n"
        pixel_string = pixel_string + "document.getElementById('badapple').innerHTML = '"
        
        # Converter o frame para escala de cinza
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Diminuir o frame para caber na tela
        image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
        
        # Quantizar o frame
        image[image >= 200] = 255
        image[image < 200] = 0
        
        # Percorrer cada pixel e adicionar à string se é preto ou branco
        for line in image:
            for pixel in line:
                if pixel == 0:
                    pixel_string = pixel_string + pixel_preto
                else:
                    pixel_string = pixel_string + pixel_branco
            
            # Pular para a próxima linha
            pixel_string = pixel_string + r"<br/>"
        
        # Passar informação do momento em que o frame é carregado
        pixel_string = pixel_string + "' \n"
        pixel_string = pixel_string + "}, "+ str(milisec) + "); \n \n"
        
        # Devolver o comando js
        return pixel_string

# Começo do arquivo content.js
javascript = ""
javascript = javascript + "chrome.runtime.onMessage.addListener(function (request) {\n"
javascript = javascript + "alert('BAD APPLE!!')\n"
javascript = javascript + "function sleep(milliseconds) {\n"
javascript = javascript + "const date = Date.now();\n"
javascript = javascript + "let currentDate = null;\n"
javascript = javascript + "do {\n"
javascript = javascript + "currentDate = Date.now();\n"
javascript = javascript + "} while (currentDate - date < milliseconds);\n"
javascript = javascript + "}\n"
javascript = javascript + "document.getElementById('badapple').style = 'line-height:60%';\n"

# Samplear o vídeo e adicionar a linha com o texto de cada frame
for s in samples:
    # P/ acompanhar a conversão...
    if s % 1000 == 0:
        print(s/10000)
    
    #getFrame(s)
    javascript = javascript + getFrame(s)

print("FIM!!")

javascript = javascript + "})"

# Salvar o script
text_file = open("content.js", "w")
text_file.write(javascript)
text_file.close()
```
